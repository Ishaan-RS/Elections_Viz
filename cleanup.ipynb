```python
import pandas as pd
```

```python
df = pd.read_csv("eci_data_2024.csv")
df
```

```python
df_t = df.copy()
df_t
```

Cleaning the data e.g. removing null ("-") value to 0

```python
df_t[["EVM Votes","Postal Votes", "Total Votes", "% of Votes"]].value_counts()
```

```python
for i in ["EVM Votes", "Postal Votes", "Total Votes", "% of Votes"]:
    df_t[i] = df_t[i].replace("-", 0)
```

```python
df_t
```

```python
df_t["Postal Votes"].value_counts()
```

```python
df = df_t
df
```

Merging the 2 files together to get the phase data in the final csv. (Phase_csv.csv was created using Phase_data.xlsx file by manual combination)

```python
de = pd.read_csv("Phase_csv.csv")
de
```

```python
dj = df.merge(de[["PC Name", "Phase"]], left_on="Constituency", right_on="PC Name")
dj
```

```python
dj["Phase"].value_counts()
```

```python
dj.to_csv("eci_data.csv", index=False)
```
